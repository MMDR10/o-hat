## What do these numbers mean?

| Balance Ratio | System Type | Real-world Example |
|:------------:|:------------|:-------------------|
| > 20× | SUPER-PULSE | 1989 Quebec geomagnetic storm |
| 5–15× | PULSE | Earthquakes M5+ |
| 1.5–4× | CONTINUOUS FLOW | ENSO, COVID-19 |
| < 1.5× | REGIME SATURATION | Seasonal forcing |

**The same three numbers work across climate, disease, earthquakes, finance, and more** — because the math doesn't care what your numbers represent.
